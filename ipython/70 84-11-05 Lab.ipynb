```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd

file = r'../data/ANALYSIS/84-11-05 lab.xls'
df = pd.read_excel(file, sheet_name=None, header=0, skiprows=[1])
df.keys()
```

```python
fig, axes = plt.subplots(nrows=12, ncols=3, figsize=(20, 45))

position = 0

for key in df.keys():
    df[key].plot(x='X-location.1', y='Y-location.1', ax=axes[position, 0])
    df[key].plot(y='Distance', ax=axes[position, 1], ylim=(0, 1.2))
    df[key].plot(y='Abs Angle', ax=axes[position, 2], ylim=(0, 180))
    position = position + 1
```

## Mean and Std Dev for Distance Moved and Abs Angle Turned

For each track what is are the summary statistics. There is clearly a high degree of variability between tracks so is it possible to compare differences between sets of tracks sampled at different times and locations.

```python
track = []
mean_dist = []
std_dist = []
mean_abs_angle = []
std_abs_angle = []

for key in df.keys():
    track.append(key)
    mean_dist.append(df[key]['Distance'].mean())
    std_dist.append(df[key]['Distance'].std())
    mean_abs_angle.append(df[key]['Abs Angle'].mean())
    std_abs_angle.append(df[key]['Abs Angle'].std())
    
summary_df = pd.DataFrame.from_items(
    [
        ('Track', track), 
        ('Mean_Dist', mean_dist), 
        ('Std_Dist', std_dist),
        ('Mean_Abs_Angle', mean_abs_angle), 
        ('Std_Abs_Angle', std_abs_angle)
    ]
)
```
