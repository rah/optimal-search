```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd

file = r'../data/ANALYSIS/84-10-19 tvl.xls'
df = pd.read_excel(file, sheetname=None, header=0, skiprows=[1])
df.keys()
```

```python
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(20, 15))

df['SNA001'].plot(x='X-location.1', y='Y-location.1', ax=axes[0, 0])
df['SNA001'].plot(y='Distance', ax=axes[0, 1])
df['SNA001'].plot(y='Abs Angle', ax=axes[0, 2])

df['SNA002'].plot(x='X-location.1', y='Y-location.1', ax=axes[1, 0])
df['SNA002'].plot(y='Distance', ax=axes[1, 1])
df['SNA002'].plot(y='Abs Angle', ax=axes[1, 2])

df['SNA003'].plot(x='X-location.1', y='Y-location.1', ax=axes[2, 0])
df['SNA003'].plot(y='Distance', ax=axes[2, 1])
df['SNA003'].plot(y='Abs Angle', ax=axes[2, 2])
```
